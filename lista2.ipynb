```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t_k = 25
c_01 = 0.2
c_02 = 0.8
c_03 = 0.3
a1 = 0.7
a2 = 1.45
a3 = 2.75
a4 = 3.2
```

```python
def logistic(c_0, t_k, a):
    c_t = [c_0]
    for i in range(t_k):
        c_t.append(a*c_t[-1]*(1-c_t[-1]))
    return c_t
```

```python
cs2 = logistic(c_02, t_k, a1)
cs1 = logistic(c_01, t_k, a1)
ts = np.linspace(0, t_k+1, t_k+1)
cs3 = logistic(c_03, t_k, a1)
plt.scatter(ts, cs1, c = "b", marker="o",label="c_0 = 0.2")
plt.plot(ts, cs1, c="b")
plt.scatter(ts, cs2, c = "r", marker="o", label="c_0 = 0.8")
plt.plot(ts, cs2, c="r")
plt.scatter(ts, cs3, c = "g", marker="o", label="c_0 = 0.3")
plt.plot(ts, cs3, c="g")
plt.legend(loc="best")
plt.title("a = 0.7")
plt.show()
```

Szukamy punktu stabilnego, a więc takiego, dla którego zachodzi
$$
c_{t+1} = c_t = c
$$
Skąd mamy 

$$
\begin{split}
c = ac(1-c) \\
c(1 - a(1-c)) = 0
\end{split}
$$
A zatem
$$
c = 0 \quad lub \quad c = \frac{a-1}{a}
$$
Aby zbadać stabilność tych punktów musimy zbadać moduł z pochodnej funkcji $f(c) = ac(1-c)$ wokół tych punktów. Mamy zatem
$$
f'(c) = a - 2ac
$$
I odpowiednio dla $c_1=0$ i $c_2=\frac{a-1}{a}$
$$
\begin{split}
|f'(c_1)| &= |a| \\
|f'(c_2)| &= |a - 2(a-1)| = |2-a|
\end{split}
$$
Teraz znajdźmy przedziały wratości $a$ dla których $|f'(c_i)| < 1$ (mamy punkt stabilny) oraz $|f'(c_i)| > 1$ (mamy punkt niestabilny).

* Dla punktu $c_1 = 0$ mamy

Stabilność gdy
$$
\begin{split}
|&a| < 1 \\
-1 < \, &a < 1
\end{split}
$$

Oraz niestabilność gdy
$$
\begin{split}
|a| &> 1 \\
a < -1 \quad &\text{lub} \quad a>1
\end{split}
$$

* Dla punktu $c_2 = \frac{a-1}{a}$ mamy

Stabilność gdy

$$
\begin{split}
|2-&a| < 1 \\
-1 < \, 2-& a < 1\\
1 < \, \,&a < 3
\end{split}
$$
Oraz niestabilność gdy
$$
\begin{split}
|2&-a| > 1 \\
1 <  2- a \quad &\text{lub} \quad   2- a < -1 \\
a < 1 \quad &\text{lub} \quad a > 3
\end{split}
$$



Wyniki omówmy na podstawie wykresów 1, 2 i 4 z poprzedniego zadania. 

Najpierw zajmijmy się sytuacją drugą, gdy $a = 1.45$. 
Z wcześniejszych rozważań wiemy, że punkt $c_1=0$ dla wartości $a > 1$ jest punktem niestabilnym, a więc takim, od którego kolejne punkty będą "szybko uciekać", gdyż dla każdego, małego wychylenia w czasie obserwujemy znaczącące zmiany kolejnych wartości. Punkt $c_2 \approx 0.31$ dla tej wartości $a$ jest z kolei punktem stabilnym, co również dobrze pokazuje wykres, ponieważ właśnie w okolicach wartości $0.31$ kumulują się kolejne punkty w czasie. Gdy trafimy na punkt stabilny, małe zaburzenia nie powodują odejścia od stanu równowagi.

Dla kontrastu, rozważmy przykład pierwszy, gdzie $a=0.7$. Wówczas, korzystając z poprzednich rozważań, wiemy, że punkt $c_1 = 0$ jest w tym wypadku punktem stabilnym. Na wykresie dobrze to widać, ponieważ kolejne punkty szybko zbliżają się do tej wartości. 

W sytuacji przedstawionej na wykresie czwartym, gdy $a=3.2$ oba punkty $c_1 = 0$ i $c_2 = 0,6875$ są punktami niestabilnymi. Widzimy, że punkty nie stabilizują się wokół żadnej z tych wartości, a wręcz wykazują pewnego rodzaju drgania wokół punktu $c_2$.

```python
def ehrenfest(k, n, p):
    fleas = [0] * n
    N_a = [n]
    N_b = [0]
    for i in range(k):
        flea_index = np.random.randint(0, n)
        u = np.random.uniform(0,1)
        if u > p:
            if fleas[flea_index] == 0:
                fleas[flea_index] = 1
            else:
                fleas[flea_index] = 0
        N_a.append(fleas.count(0))
        N_b.append(fleas.count(1))
    return N_a, N_b

```

```python
k = 10000
n1 = 1000
n2 = 100
n3 = 10000
p1 = 0.5
p2 = 0.1
p3 = 0.9
```

```python
ts = np.linspace(0, k+1, k+1)
d1, d2 = ehrenfest(k, n1, p1)
d1_1, d2_1 = ehrenfest(k, n2, p1)
d1_2, d2_2 = ehrenfest(k, n3, p1)
fig, axis = plt.subplots(3,1)
axis[0].plot(ts, d1, c="b", label="pies A")
axis[0].plot(ts, d2, c="r", label="pies B")
axis[0].set_title("Liczba pcheł: 1000, p = 0.5")
axis[1].plot(ts, d1_1, c="b", label="pies A")
axis[1].plot(ts, d2_1, c="r", label="pies B")
axis[1].set_title("Liczba pcheł: 100, p = 0.5")
axis[2].plot(ts, d1_2, c="b", label="pies A")
axis[2].plot(ts, d2_2, c="r", label="pies B")
axis[2].set_title("Liczba pcheł: 10000, p = 0.5")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python

d3, d4 = ehrenfest(k, n1, p2)
d3_1, d4_1 = ehrenfest(k, n2, p2)
d3_2, d4_2 = ehrenfest(k, n3, p2)
fig, axis = plt.subplots(3,1)
axis[0].plot(ts, d3, c="b", label="pies A")
axis[0].plot(ts, d4, c="r", label="pies B")
axis[0].set_title("Liczba pcheł: 1000, p = 0.1")
axis[1].plot(ts, d3_1, c="b", label="pies A")
axis[1].plot(ts, d4_1, c="r", label="pies B")
axis[1].set_title("Liczba pcheł: 100, p = 0.1")
axis[2].plot(ts, d3_2, c="b", label="pies A")
axis[2].plot(ts, d4_2, c="r", label="pies B")
axis[2].set_title("Liczba pcheł: 10000, p = 0.1")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python

d5, d6 = ehrenfest(k, n1, p3)
d5_1, d6_1 = ehrenfest(k, n2, p3)
d5_2, d6_2 = ehrenfest(k, n3, p3)
fig, axis = plt.subplots(3,1)
axis[0].plot(ts, d5, c="b", label="pies A")
axis[0].plot(ts, d6, c="r", label="pies B")
axis[0].set_title("Liczba pcheł: 1000, p = 0.9")
axis[1].plot(ts, d5_1, c="b", label="pies A")
axis[1].plot(ts, d6_1, c="r", label="pies B")
axis[1].set_title("Liczba pcheł: 100, p = 0.9")
axis[2].plot(ts, d5_2, c="b", label="pies A")
axis[2].plot(ts, d6_2, c="r", label="pies B")
axis[2].set_title("Liczba pcheł: 10000, p = 0.9")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

Z wykresów widzimy, że zwiększanie liczby pcheł powoduje wydłużenie się czasu dojścia do stanu równowagi w liczbie pcheł na każdym psie. Jest to dość intuicyjna obserwacja, ponieważ im więcej pcheł początkowo znajdzie się na psie A, tym więcej czasu będziemy potrzebowali na wyrównanie się wartości $N_a$ i $N_b$.       Zmniejszenie liczebności pcheł powoduje szybkie wyrównanie się wartości $N_a$ i $N_b$, a nastepujące w kolejnych chwilach czasu zmiany, wykazują fluktuacje wokół położenia równowagi, gdzie stosunek $N_a$ do $N_b$ wynosi około 1:1.

Podobna sytuacja ma miejsce w przypadku zwiększania parametru $p$. Możemy to wytłumaczyć tym, że im większa jest ruchliwość pcheł, tym częściej będą przeskakiwać między psami, i tym samym, stan równowagi będzie trudniejszy do osiągnięcia (będzie potrzebne więcej czasu do jego osiągnięcia).
